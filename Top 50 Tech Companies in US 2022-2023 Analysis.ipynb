**Import Libraries and Dataset**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tech_com = pd.read_csv('Top 50 US Tech Companies 2022 - 2023.csv')
```

```python
tech_com.head()
```

```python
tech_com.info()
```

```python
tech_com.isnull().sum()
```

---

**Amount of company sectors**

```python
tech_com['Sector'].value_counts()
```

```python
plt.figure(figsize=(20,10))
plt.title("The most technology company sector")
sns.countplot(x='Sector', data=tech_com, order=tech_com['Sector'].value_counts().index)
plt.xlabel("Sector")
plt.ylabel("Count of Sector")
```

**Amount of Headquarter States**

```python
tech_com['HQ State'].value_counts()
```

```python
plt.figure(figsize=(18,8))
plt.title("The most technology company in states")
sns.countplot(x='HQ State', data=tech_com, order=tech_com['HQ State'].value_counts().index)
plt.xlabel("State")
plt.ylabel("Count of State")
```

**Top 10 highest annual revenue**

```python
top_anrev = tech_com[['Company Name', 'Annual Revenue 2022-2023 (USD in Billions)']].nlargest(10,'Annual Revenue 2022-2023 (USD in Billions)')
top_anrev
```

```python
plt.figure(figsize=(20,8))
plt.title("Top 10 Highest Tech Company Annual Revenue Between 2022 to 2023")
sns.barplot(x=top_anrev['Company Name'], y=top_anrev['Annual Revenue 2022-2023 (USD in Billions)'])
plt.xlabel("Company")
plt.ylabel("Annual Revenue (USD in Billions)")
```

**Top 10 highest market cap**

```python
top_mar_cap = tech_com[['Company Name', 'Market Cap (USD in Trillions)']].nlargest(10,'Market Cap (USD in Trillions)')
top_mar_cap
```

```python
plt.figure(figsize=(24,8))
plt.title("Top 10 Highest Tech Company Market Cap Between 2022 to 2023")
sns.barplot(x=top_mar_cap['Company Name'], y=top_mar_cap['Market Cap (USD in Trillions)'])
plt.xlabel("Company")
plt.ylabel("Market Cap (USD in Trillions)")
```

**Top 10 highest annual income tax**

```python
top_ann_tax = tech_com[['Company Name', 'Annual Income Tax in 2022-2023 (USD in Billions)']]\
.nlargest(10,'Annual Income Tax in 2022-2023 (USD in Billions)')
top_ann_tax
```

```python
plt.figure(figsize=(24,10))
plt.title("Top 10 Highest Tech Company Annual Income Tax Between 2022 to 2023")
sns.barplot(x=top_ann_tax['Company Name'], y=top_ann_tax['Annual Income Tax in 2022-2023 (USD in Billions)'])
plt.xlabel("Company")
plt.ylabel("Annual Income Tax (USD in Billions)")
```

**Newest Technology Company**

```python
tech_com[tech_com['Founding Year'] == tech_com['Founding Year'].max()]
```

**Oldest Technology Company**

```python
tech_com[tech_com['Founding Year'] == tech_com['Founding Year'].min()]
```

**Top 10 highest number of employees in each company**

```python
top_emp = tech_com[['Company Name', 'Employee Size']].nlargest(10,'Employee Size')
top_emp
```

```python
plt.figure(figsize=(20,8))
plt.title("Top 10 Highest Employee Amount")
sns.barplot(x=top_emp['Company Name'], y=top_emp['Employee Size'])
plt.xlabel("Company")
plt.ylabel("Number of Employee")
```
